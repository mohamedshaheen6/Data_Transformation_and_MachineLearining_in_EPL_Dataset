```python
import pandas as  pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import preprocessing
```

```python
epl=pd.read_csv("C:/Users/ASUS/Desktop/DS SM/ML/Project/eplmatches.csv")
epl
```

```python
awaygame=epl.iloc[:,[6,4,5,7]]
awaygame
```

```python
awaygamescore=awaygame.iloc[:,1:3]
awaygamescore
```

```python
wcss=[]
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=1)
model=kmeans.fit_predict(awaygamescore)
model
```

```python
model_series=pd.Series(model)
model_series
```

```python
awaygamescore['Cluster']=model_series
awaygamescore
```

```python
awaygamescore.groupby(awaygamescore.Cluster).mean()
```
